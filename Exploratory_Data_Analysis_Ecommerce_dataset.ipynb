```python
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
df['Gender'].value_counts().plot(kind='bar', color='skyblue', title='Gender Distribution')
plt.ylabel('Count')

plt.subplot(2, 2, 2)
df['Age Group'].value_counts().plot(kind='bar', color='lightgreen', title='Age Group Distribution')
plt.ylabel('Count')

plt.subplot(2, 2, 3)
df['Product Category'].value_counts().plot(kind='bar', color='lightcoral', title='Product Category Distribution')
plt.ylabel('Count')

plt.subplot(2, 2, 4)
df['Purchase Method'].value_counts().plot(kind='bar', color='orange', title='Payment Method Distribution')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Convert 'Purchase Date' to datetime if it's not already
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

# Now you can use the .dt accessor
df['purchase_year_month'] = df['Purchase Date'].dt.to_period('M')
```

```python
purchase_trend = df.groupby('purchase_year_month').size()
```

```python
purchase_trend
```

```python
plt.figure(figsize=(12, 6))
purchase_trend.plot(kind='line', marker='o', color='blue', title='Purchase Trend Over Time')
plt.xlabel('Purchase Year-Month')
plt.ylabel('Number of Purchases')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
